#### 6. Write a program to check the data type of the entered values.

```python
x = input("Enter the value: ")

try:
    if type(eval(x)) == float:
        print ("The data type of the input value is: float")
    elif type(eval(x)) == int:
        print ("The data type of the input value is: integer")    
    elif type(eval(x)) == bool:
        print ("The data type of the input value is: Boolean")      
except:
    print ("The data type of the input value is: string")
```

#### 7. Create Variables using formats such as Upper CamelCase, Lower CamelCase, SnakeCase and UPPERCASE.

```python
word = input(" ")

def camel(s):
    
    s = "".join(word.title().split())
    return ''.join([s[0].lower(), s[1:]])

def camel_upper(s):
    s = "".join(word.title().split())
    return ''.join([s[0].upper(), s[1:]])

def to_snake_case(value):
    return "_".join(value.lower().split())


result = word.upper()
print(" Upper case : {}".format(result))
result2 = camel(word)
print(" lower camel case : {}".format(result2))

result3 = camel_upper(word)
print(" upper camel case : {}".format(result3))


result4 = to_snake_case(word)
print(" snake case : {}".format(result4))
```

#### 10. Write a program that asks five times to guess the lucky number. Use a while loop and a counter

```python
counter = 1
lucky_number = 8

while counter <= 5:
    counter = counter + 1
    user_input = int(input('Guess the number: '))
    if user_input == lucky_number:
        print('Good guess!\n')
        
    else:
        print('Try again!\n')
print('Game over')
```

#### 11. In the previous question, insert break after the “Good guess!” print statement. break will terminate the while loop so that users do not have to continue guessing after they found the number. If the user does not guess the number at all, print “Sorry but that was not very successful”.

```python
counter = 1
lucky_number = 8
non_successful = False
 
while counter <= 5:
    counter = counter + 1
    user_input = int(input('Guess the number: '))
    if user_input == lucky_number:
        print('Good guess!\n')
        break
    else:
        print('Try again!\n')
        if counter == 5:
            non_successful = True
        
if non_successful is True:
    print('Sorry but that was not very successful')  
else:
    print('Game over')
```
